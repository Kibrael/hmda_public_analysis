```python
#Race/Action matrix counts

sql = """
SELECT * 
FROM crosstab( 
'SELECT action_taken_type, race1, count(race1)
from {table} group by 1,2 order by 1,2')  
AS final_result(
"action_type" varchar, 
"native" bigint, "asian" bigint, "black" bigint, "islander" bigint, "white" bigint, "no_info" bigint
,"NA" bigint)"""
```

```python
#action frequency, percent, cumulative percent
def get_action_frequency(table):
    sql = """
    SELECT '2018' AS year, action_taken_type
    ,(ROUND(COUNT(action_taken_type) *100.0/ (SELECT COUNT(*) FROM {table}) )) AS freq_pct
    ,(CASE WHEN action_taken_type='1' THEN COUNT(action_taken_type) 
    WHEN action_taken_type='2' THEN COUNT(action_taken_type)  
    WHEN action_taken_type='3' THEN COUNT(action_taken_type) 
    WHEN action_taken_type='4' THEN COUNT(action_taken_type) 
    WHEN action_taken_type='5' THEN COUNT(action_taken_type) 
    WHEN action_taken_type='6' THEN COUNT(action_taken_type) 
    WHEN action_taken_type='7' THEN COUNT(action_taken_type)  
    WHEN action_taken_type='8' THEN COUNT(action_taken_type) END) AS action_count
    FROM {table}
    GROUP BY action_taken_type
    ORDER BY action_taken_type""".format(table=table)
    pg_conn.execute(sql)
    colnames = [desc[0] for desc in pg_conn.description]
    data_df = pd.DataFrame(pg_conn.fetchall(), columns=colnames)
    return data_df

first = True
pg_conn = connect()
for table in prod_tables:
    action_df = get_action_frequency(table)
    print(action_df.head())
    if first:
        first = False
        action_freq_df = action_df.copy()
    else:
        action_freq_df = pd.concat([action_freq_df, action_df])
        
pg_conn.close()
```
